# Feature Engineering

```python
import numpy as np

# Recreate features from original columns
df['Log_Amount'] = np.log1p(df['Amount'])  # Log transform for 'Amount'
df['Hour'] = (df['Time'] // 3600) % 24      # Extract 'Hour' from 'Time'
df['Day_Night'] = np.where(df['Hour'].isin(range(7, 19)), 1, 0)  # 1 = Day, 0 = Night
```

```python
df['Amount_per_Hour'] = df.groupby('Hour')['Amount'].transform('mean')
```

```python
df['Amount_vs_Time'] = df['Amount'] / (df['Time'] + 1)  
```

```python
import numpy as np
import pandas as pd

# Load dataset
data = pd.read_csv("creditcard.csv")

# Create time-based series for FFT analysis
data['Time'] = pd.to_datetime(data['Time'], unit='s')
data.set_index('Time', inplace=True)

# Resample 'amount' data in 1-minute intervals for time-series behavior
amount_series = data['Amount'].resample('1T').sum().fillna(0)

# FFT Feature Extraction Function
def extract_fft_features(signal):
    fft_values = np.fft.fft(signal)
    fft_magnitudes = np.abs(fft_values)
    
    return {
        'Max_Freq_Amp': np.max(fft_magnitudes),
        'Top_5_Freqs': np.sort(fft_magnitudes)[-5:].sum(),
        'Energy': np.sum(fft_magnitudes ** 2) / len(fft_magnitudes)
    }

# Extract FFT features
fft_features = extract_fft_features(amount_series)

# Combine FFT features with the original dataset
fft_features_df = pd.DataFrame([fft_features])
final_df = pd.concat([data.reset_index(), fft_features_df], axis=1)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.combine import SMOTETomek
from sklearn.preprocessing import StandardScaler

# Load dataset
data = pd.read_csv("creditcard.csv")

# Splitting data into features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Standardizing features to improve SMOTE performance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply Borderline-SMOTE for better minority sampling
smote = BorderlineSMOTE(sampling_strategy=0.3, random_state=42)  # Reduced ratio
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

# Combine SMOTE with undersampling (less aggressive strategy)
smote_tomek = SMOTETomek(sampling_strategy='auto', random_state=42)  # Auto balances classes
X_train_final, y_train_final = smote_tomek.fit_resample(X_train_smote, y_train_smote)

# Train a RandomForest model with class weighting
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=20,
    class_weight='balanced',
    random_state=42
)

# Fit the model
model.fit(X_train_final, y_train_final)

# Predictions and evaluation
y_pred = model.predict(X_test_scaled)
print(classification_report(y_test, y_pred))

```
